```python
# Функция для красивого вывода ингредиентов на экран.
def print_ingredients(ingredients):
	print("Яйца:", ingredients[0], "шт.")
	print("Мука:", ingredients[1], "ст.")
	print("Молоко:", ingredients[2], "л")
	print("Сахар:", ingredients[3], "ст. л.")
	print("Соль:", ingredients[4], "ч. л.")
	print("Растительное масло:", ingredients[5], "ст. л.")
	print()
```

```python
# пример данных
ingredients = [2, 3, 1, 2, 0.5, 1]

# вызов функции
print_ingredients(ingredients)
```

```python
def calc_ingredients(n_portions):
    recipe = np.array([3/4, 1/4, 0.4/4, 1/4, 0.5/4, 1.5/4])
    product = recipe * n_portions

    return product
```

```python
# Пример использования функции print_ingredients
ingredients_5 = calc_ingredients(5)
print_ingredients(ingredients_5)
print()
ingredients_8 = calc_ingredients(8)
print_ingredients(ingredients_8)
print()
ingredients_10 = calc_ingredients(10)
print_ingredients(ingredients_10)
```

```python
# Воспользуйтесь векторами для прогноза сразу для всех месяцев.
import numpy as np

start = 3210

def y(start, n):
    forecast = start + 130 * n
    return forecast

may = 5
august = 8
september = 9
december = 12

forecast_may = y(start, may)
forecast_august = y(start, august)
forecast_september = y(start, september)
forecast_december = y(start, december)

print(f"End of May: {forecast_may} subscribers")
print(f"End of August: {forecast_august} subscribers")
print(f"End of September: {forecast_september} subscribers")
print(f"End of December: {forecast_december} subscribers")
```

```python
# y = -2x^2 + 14x + 1
a, b, c = -2, 14, 1

# x координата вершины
x_max = -b / (2 * a)

# максимальное значение y
y_max = a * x_max**2 + b * x_max + c

print(f"x at max height: {x_max}")
print(f"Max height: {y_max}")
```

```python
import numpy as np

def Q(x):
    return 5000 - 6.25 * x

def P(x):
    return x * Q(x) - (30000 + 120 * Q(x))

# максимум параболы: P(x) = -6.25x^2 + 5750x - 630000
a = -6.25
b = 5750
x_opt = -b / (2 * a)

q_opt = Q(x_opt)
p_opt = P(x_opt)

print("x_opt:", x_opt)
print("Q(x_opt):", q_opt)
print("Max profit:", p_opt)
```

```python
import numpy as np

# Система для y = a*x^2 + b*x + c
# Подставляем точки: (1, -2), (-4, 3), (0.5, 0.75)

A = np.array([
    [1**2,   1,   1],
    [(-4)**2, -4,  1],
    [(0.5)**2, 0.5, 1]
], dtype=float)

Y = np.array([-2, 3, 0.75], dtype=float)

a, b, c = np.linalg.solve(A, Y)

print(f"a = {a}")
print(f"b = {b}")
print(f"c = {c}")
print(f"y = {a}x^2 + {b}x + {c}")
```

```python
import numpy as np

# Матрица для системы: a*x^3 + b*x^2 + c*x + d = y
A = np.array([
    [0**3, 0**2, 0, 1],     # x=0
    [(-1)**3, (-1)**2, -1, 1],  # x=-1
    [(-4)**3, (-4)**2, -4, 1],  # x=-4
    [1**3, 1**2, 1, 1]      # x=1
], dtype=float)

Y = np.array([-1, 0, 3, -12], dtype=float)

a, b, c, d = np.linalg.solve(A, Y)

print("a =", a)
print("b =", b)
print("c =", c)
print("d =", d)
```

```python
import numpy as np

t = 22  # years after 2000 (t=1 for 2001, so t=22 for 2022)

C = 0.17 * t**3 - 2.58 * t**2 + 24.75 * t - 1.39

print("C (thousand EVs) in 2022:", C)
```

```python
import numpy as np

# polynomial: C(t) = 0.17 t^3 - 2.58 t^2 + 24.75 t - 1.39
C_poly = np.poly1d([0.17, -2.58, 24.75, -1.39])  # coefficients from highest power to constant

t = 22  # 2022 year => 22 years after 2000

C_2022 = C_poly(t)

print(C_2022)
```
