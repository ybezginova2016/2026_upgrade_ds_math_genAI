```python
# 1) Задаем два вектора s и w (наборы чисел одной длины)
s = np.array([5, 3.1, 8, 7, 0])
w = np.array([13, 2.6, 4, 9, 7])

# 2) Считаем скалярное произведение s · w
#    dot = 5*13 + 3.1*2.6 + 8*4 + 7*9 + 0*7
dot_sw = np.dot(s, w)

# 3) Считаем евклидовы нормы (длины) векторов |s| и |w|
#    |s| = sqrt(5^2 + 3.1^2 + 8^2 + 7^2 + 0^2)
#    |w| = sqrt(13^2 + 2.6^2 + 4^2 + 9^2 + 7^2)
norm_s = np.linalg.norm(s)
norm_w = np.linalg.norm(w)

# 4) Считаем косинус угла между векторами по формуле:
#    cos(theta) = (s · w) / (|s| * |w|)
cos_angle = dot_sw / (norm_s * norm_w)

# 5) Из-за округлений в вычислениях cos(theta) может получиться чуть больше 1
#    или чуть меньше -1, и тогда arccos даст ошибк
```

```python
import numpy as np

# 1) Координаты вершин четырехугольника BADC
B = np.array([-15, 15])
A = np.array([-4, 8])
D = np.array([3, -4])
C = np.array([-10, -7])

# 2) Угол D лежит между векторами DA и DC (оба выходят из точки D)
DA = A - D
DC = C - D

# 3) Скалярное произведение DA · DC
dot_product = np.dot(DA, DC)

# 4) Нормы (длины) векторов |DA| и |DC|
norm_DA = np.linalg.norm(DA)
norm_DC = np.linalg.norm(DC)

# 5) Косинус угла D по формуле:
#    cos(D) = (DA · DC) / (|DA| * |DC|)
cos_angle = dot_product / (norm_DA * norm_DC)

# 6) Защита от погрешностей округления: значение должно быть строго в [-1, 1]
cos_angle = np.clip(cos_angle, -1.0, 1.0)

# 7) Находим угол в радианах через arccos
angle_rad = np.arccos(cos_angle)

# 8) Переводим угол в градусы
angle_deg = np.degrees(angle_rad)

# 9) Выводим значения в нужном формате
print("Косинус угла:", cos_angle)
print("Сам угол:", angle_deg)
```

```python
import numpy as np

# daily difference vs target (EUR)

diff = np.array([120, -80, 50, -30, 200])

l1 = np.linalg.norm(diff, 1)
# total deviation across the week

l2 = np.linalg.norm(diff, 2)
# typical deviation (penalizes larger spikes)

linf = np.linalg.norm(diff, np.inf)
# worst day deviation

print("L1 total deviation:", l1)
print("L2 typical deviation:", l2)
print("L∞ worst day:", linf)
```

```python
import numpy as np

# Typical shipment profiles (centroids)
# [weight_kg, pallets, lead_time_days, documents]
standard = np.array([800, 12, 10, 1])
urgent   = np.array([300,  4,  3, 1])
complex  = np.array([1200, 20, 14, 3])

centroids = np.array([standard, urgent, complex])
labels = ["STANDARD", "URGENT", "COMPLEX"]

# New shipment to classify
x_new = np.array([950, 15, 12, 2])

# Euclidean distance to each centroid
dist = np.linalg.norm(centroids - x_new, axis=1)

print("Distances:", dist)
print("Shipment type:", labels[np.argmin(dist)])
```
