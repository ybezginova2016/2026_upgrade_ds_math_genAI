## Инструменты описания данных в Python

```python
import numpy as np
import matplotlib.pyplot as plt

# Дни, для которых нужно посчитать N(t)
days = np.array([1, 5, 10, 20, 30])

# Вектор t для построения графика (0..70 с шагом 0.1)
x = np.arange(0, 70, 0.1)

# Функция производительности N(t)
def N(t):
    return 80 - 70 * (2.72 ** (-0.13 * t))

# 1) Считаем N(t) для дней 1, 5, 10, 20, 30
result = N(days)

print("N(t) for days [1, 5, 10, 20, 30]:", result)

# 2) Строим график N(t)
plt.plot(x, N(x))
plt.grid()
plt.xlabel("Days since start (t)")
plt.ylabel("Phones per day N(t)")
plt.title("Employee productivity growth N(t)")
plt.show()

# 3) Потолок эффективности (предел функции при t -> infinity)
# Когда t растет, степень становится очень маленькой -> выражение стремится к 0
# Тогда N(t) -> 80
ceiling = 80

print("Ceiling productivity (phones/day):", ceiling)
```

```python
import numpy as np
import matplotlib.pyplot as plt

days = np.array([1, 5, 10, 20, 30])
x = np.arange(0, 70, 0.1)

def N(t):
    return 80 - 70 * (2.72 ** (-0.13 * t))

# 1) значения для 1, 5, 10, 20, 30 дней
result = N(days)
print("N(t) for days [1, 5, 10, 20, 30]:", result)

# 2) график
plt.plot(x, N(x))
plt.grid()
plt.show()

# 3) "потолок" по графику: берём значение в конце интервала (t=70)
ceiling = int(round(N(70)))
print("Ceiling productivity (phones/day):", ceiling)
```
