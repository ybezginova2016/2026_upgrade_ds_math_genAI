### L1 (Manhattan) distance

For two vectors
$\mathbf{a} = (a_1, a_2, \dots, a_n)$ and
$\mathbf{b} = (b_1, b_2, \dots, b_n)$

the **L1 distance** is defined as:

$$
L_1(\mathbf{a}, \mathbf{b}) = \sum_{i=1}^{n} |a_i - b_i|
$$

Идея решения (словами):
- Вычесть один вектор из другого по координатам
- Взять модуль разностей координат
- Сложить полученные значения
- Это и есть формула L1-расстояния.

```python
def l1_dist (a, b):
    manh_diff = 0
    for i in range(len(a)):
        manh_diff += abs(a[i] - b[i])
    return manh_diff
```

### Формула L2 (евклидово расстояние)

Для двух векторов
\(\mathbf{a} = (a_1, a_2, \dots, a_n)\) и
\(\mathbf{b} = (b_1, b_2, \dots, b_n)\)

евклидово расстояние определяется по формуле:

$$
L_2(\mathbf{a}, \mathbf{b}) =
\sqrt{
\sum_{i=1}^{n} (a_i - b_i)^2
}
$$


```python
def l2_sq(a, b):
    total_l2_sq = 0
    for i in range(len(a)):
        total_l2_sq += (a[i] - b[i]) ** 2

    return total_l2_sq
```

```python
a = array([22, 4, -6, 3.9, 7])
b = array([12, -0.5, 6, 42, 1])

r_1 = norm(a - b, 1)  # Вычисление L1 расстояния.
r_2 = norm(a - b, 2)  # Вычисление L2 расстояния.
r_max = norm(a - b, inf)  # Вычисление максимального расстояния.
print("L1 расстояние между векторами a и b:", r_1)  # Вывод на экран L1 расстояния.
print("L2 расстояние между векторами a и b:", r_2)  # Вывод на экран L2 расстояния.
print(
    "Максимальное расстояние между векторами a и b:", r_max
)  # Вывод на экран максимального расстояния.
```

```python
def l_max (a, b):
    max_diff = 0
    for ai, bi in zip(a, b):
        diff = abs(ai - bi)
        if diff > max_diff:
            max_diff = diff
    return max_diff
```

```python
def cosine_similarity(a, b):
    return a @ b / (norm(a) * norm(b))
```

```python
print("ГорСМИ и Новости городка:", cosine_similarity(v1, v2))
print("ГорСМИ и Природный вестник:", cosine_similarity(v2, v3))
print("Новости городка и Природный вестник:", cosine_similarity(v1, v3))
```
