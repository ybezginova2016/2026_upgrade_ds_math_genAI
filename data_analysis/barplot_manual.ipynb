```python
import seaborn as sns # импортируем библиотеку seaborn
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12, 6) # указываем размер визуализации
```

```python
# График 1
# создаём столбец years с годами выпуска игр с помощью метода pandas.Series()
years = pd.Series([1997, 1998, 1999, 2000, 2001, 2002,
         2003, 2004, 2005, 2006, 2007, 2008,
         2009, 2010, 2011, 2012, 2013, 2014,
         2015, 2016, 2017, 2018, 2019])

# создаём столбец games_released с количеством выпущенных игр по годам
# в нём столько же элементов, сколько в years
games_released = pd.Series([1, 1, 2, 2, 4, 1, 3, 6, 6, 48, 93, 145, 305,
                  238, 239, 320, 418, 1555, 2597,
                  4361, 6357, 8160, 2213])

sns.barplot(x=years, y=games_released) # вызываем функцию barplot()

# значения years будут на оси X
# значения games_released — на оси Y
```
