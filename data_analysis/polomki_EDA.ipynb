### 1. Среднее значение по неделям (главная динамика)

```python
means = polomki.mean()

plt.figure()
plt.plot(means.index, means.values, marker='o')
plt.xticks(rotation=45)
plt.title("Average value by week")
plt.xlabel("Week")
plt.ylabel("Average value")
plt.grid()
plt.show()
```

### 2. Разброс значений (нестабильность)

```python
stds = polomki.std()

plt.figure()
plt.plot(stds.index, stds.values, marker='o')
plt.xticks(rotation=45)
plt.title("Standard deviation by week")
plt.xlabel("Week")
plt.ylabel("Std deviation")
plt.grid()
plt.show()
```

### 3. Boxplot для отчёта

```python
plt.figure(figsize=(12,5))
polomki.boxplot(rot=45)
plt.title("Distribution by week")
plt.show()
```

## Общий график: ***mean + разброс***

```python
# Средние значения и разброс
means = polomki.mean()
stds = polomki.std()

plt.figure(figsize=(10, 5))

# Линия среднего
plt.plot(means.index, means.values, marker="o", label="Average")

# Зона разброса
plt.fill_between(
    means.index,
    means - stds,
    means + stds,
    alpha=0.2,
    label="Std range"
)

plt.title("Breakdowns dynamics by week")
plt.xlabel("Week")
plt.ylabel("Average breakdown level")
plt.xticks(rotation=45)
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

## Вывод по графику поломок

1. **Общий тренд растущий**
   Средний уровень поломок постепенно увеличивается с 1-й по примерно 10–12-ю неделю.

2. **Пик поломок наблюдается около 10-й недели**
   Именно в этот период достигаются максимальные значения.

3. **После пика наблюдается небольшое снижение**, но уровень остаётся выше, чем в начале периода.

4. **Разброс между магазинами увеличивается к середине периода**
   Заштрихованная область показывает рост различий между магазинами — часть точек работает хуже остальных.

5. **Резкое падение на последней неделе**
   Значение на 14-й неделе близко к нулю, что, вероятно, связано с неполными данными или незавершённым периодом отчёта.

## Бизнес-вывод для отчёта (на русском)

Количество поломок постепенно растёт к середине анализируемого периода, достигая максимума примерно на 10–12 неделях. Одновременно увеличивается различие между магазинами, что указывает на неравномерное качество работы оборудования или процессов в разных точках. Резкое снижение на последней неделе, скорее всего, связано с неполнотой данных.

## Business conclusion for report (English)
