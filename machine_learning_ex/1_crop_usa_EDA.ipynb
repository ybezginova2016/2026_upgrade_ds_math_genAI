# Crop Yield Data (1985-2019)

### Dataset Description

The dataset contains agricultural statistics from the **U.S. Department of Agriculture** covering the period **1980–2019**.

Each record represents crop production data for a specific **state** and **year**.

### Dataset structure
- **Year** — reporting year.
- **State** — U.S. state name.
- **State ANSI** — numerical code of the state.
- **Acres** — planted area in acres.
- **Production** — crop production volume in bushels.

### Additional metric
Crop **yield** is not provided directly but can be calculated as:

Yield = Production / Acres

The dataset therefore allows analysis of:
- changes in planted area,
- crop production dynamics,
- yield trends across states over time.

### Logically, where there are larger cultivated areas, there should be larger harvests. However, there are nuances. For example, in Texas, the harvest in 2019 was approximately the same as in South Dakota, even though South Dakota has about three times less cultivated land.

```python
# 1) group the data by year (Year) using groupby('Year')
# 2) select from the grouped table only the column with cultivated area
# 3) sum the values in each group using sum()
acres_usa = crops_usa.groupby('Year')['Acres'].sum()
print(acres_usa)
```

```python
acres_usa = crops_usa.groupby('Year')['Acres'].sum()
acres_usa
```

```python
production_usa = crops_usa.groupby('Year')['Production'].sum()

# делим значения столбца production_usa на значения столбца acres_usa
yield_usa = production_usa / acres_usa
print(yield_usa)
```

```python
# 1) extract the index values (unique years) from the Series acres_usa
# 2) the index represents calendar years, ranging from 1980 to 2019
years_unique = acres_usa.index.values

years_unique
```

```python
# стром столбчатую диаграмму урожайности по годам
sns.barplot(x=years_unique, y=yield_usa)
plt.show()
```
