# Tuples: Nesting

```python
NT = (1, 2, ("pop", "rock"), (3,4), ("disco", (1,2)))
NT
```

```python
NT[0]
```

```python
NT[2]        # ("pop", "rock")
NT[2][0]     # "pop"
NT[2][1]     # "rock"

NT[4]        # ("disco", (1, 2))
NT[4][0]     # "disco"
NT[4][1]     # (1, 2)
NT[4][1][0]  # 1
NT[4][1][1]  # 2
```

```python
NT = (1, 2, ("pop", "rock"), (3, 4), ("disco", (1, 2)))
NT_new = NT[:3] + (("hiphop", "rock"),) + NT[4:]
print(NT_new)
```

```python
sorted(c_tuple) # меняем и создаем НОВЫЙ список

# sorted() - это функция: Она работает почти с любым итерируемым объектом и возвращает новый отсортированный список.

# .sort() - это метод список: Он сортирует список на месте (in-place), то есть меняет сам объект.
```

```python
a = [3, 1, 2] # меняем САМ объект, новый список НЕ создается
a.sort()
print(a)  # [1, 2, 3]
```

```python
# Важно: .sort() возвращает None:

a = [3, 1, 2]
result = a.sort()
print(result)  # None
```
